```python
def GetCenterElement(numarray):
    if (numarray.size % 2)==0:
        x=int(numarray.size / 2)
        print("first element:" + str(numarray[x-1]) + " and second element:" + str(numarray[x]))
    else:
        x=int(numarray.size / 2)
        print("first element:" + str(numarray[x]))


arr1 = np.array([100, 103,104,109,111]);
arr2= np.array([223,222,100,101]);

GetCenterElement(arr1);
GetCenterElement(arr2);
```

# 37 Write a function to print all the upper case letter of a given sentence.
